```python
# get all names of pokemon
import time
pokemons = []
pokemon_game_info = "https://pokeapi.co/api/v2/pokemon/"
counter = 0
for pokemon in paldea_pokemon:
    try:
        pokemon_name = pokemon["pokemon_species"]["name"]
        pokemon_url = pokemon["pokemon_species"]["url"]
        # should really use pokemon id here instead of pokemon name. allows for the actual pokemon to appear if they have forms
        pokemon_properties = requests.get(pokemon_game_info + pokemon_name)

        pokemon_properties = pokemon_properties.json()
        pokemon_moves = pokemon_properties["moves"]
        pokemon_types = pokemon_properties["types"]
        pokemon_abilities = pokemon_properties["abilities"]
        pokemon_stats = pokemon_properties["stats"]

        pokemons.append({
            "name": pokemon_name,
            "moves": pokemon_moves,
            "types": pokemon_types,
            "abilities": pokemon_abilities,
            "stats": pokemon_stats,
            # should include id here
            "id": pokemon_properties["id"]
        })
        counter = counter+1
    except:
        print("Waiting.... some weird error occured")
        print(pokemon_name)
        






```

```python
#get all types

type_relations = {}

for t in range(1, 19):
    root_type_url = "https://pokeapi.co/api/v2/type/"
    type_info = requests.get(root_type_url+str(t)).json()

    type_relations[type_info["name"]] = type_info["damage_relations"]
```

```python
sv_moves = []

for poke in pokemons:
    moves = poke["moves"]
    for move in moves:
        sv_moves.append({
            "move": move["move"]["name"],
            "url": move["move"]["url"]
        })
```

```python
import pandas as pd
move_df = pd.DataFrame(sv_moves)
```

```python
move_df = move_df.drop_duplicates()
```

```python
def extract_move_properties(url):
    # get properties given url
    result = requests.get(url).json()
    return {
        "accuracy": result["accuracy"],
        "category": result["damage_class"],
        "id": result["id"],
        "bp": result["power"],
        "type": result["type"]["name"],
        "name": result["name"]
    }

    
```

```python
move_properties = move_df.url.apply(lambda x: extract_move_properties(x))
```

```python
moves = pd.DataFrame(move_properties.tolist())[["name", "accuracy", "id", "bp", "type"]]
# replace nans with zeros
```

```python
import re
def search_json_dict(string):
    # thx chatgpt
    pattern = r'{.*?}'
    match = re.search(pattern, string)
    if match:
        json_dict_string = match.group()
        return json_dict_string
    else:
        return None
```

```python
string = '''Input: Tell me if max speed iron bundle outspeeds max speed iron hands

Output:

{ "p1": "Iron Bundle", "p2": "Hands", "p1_stat_changes": 0, "p2_stat_changes": 0, "p1_ev": 252, "p2_ev": 252 }'''

import re
import ast
ast.literal_eval(search_json_dict(string).replace(" ", ""))
```

```python

def regex_after_keyword(string, keyword):
    pattern = r'(?<={}).*'.format(re.escape(keyword))
    match = re.search(pattern, string)
    if match:
        result = match.group()
        return result
    else:
        return None

regex_after_keyword(string.replace(" ", ""), "Output")
```
